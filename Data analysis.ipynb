```r
# Make copy of dataframe
train2 <- data.frame(train)
```

```r
# Remove Id and Class attributes
train2 <- train2[ , -which(names(train2) %in% c("Id","Class"))]
```

```r
head(train2,5)
```

```r
# List of conditions
conditions <- train$Class
conditions
```

```r
# Get the list of symptoms
symptoms <- train2
head(symptoms)
```

```r
# Dictionary that maps symptoms to conditions
symptom_to_condition <- setNames(conditions, symptoms)
plot(symptom_to_condition)
```

```r
train2$symptom_to_condition <- symptom_to_condition
```

```r
ggplot(train2, aes(x = symptom_to_condition)) +
  geom_histogram(color = "Red", fill = "white", size = 1.5) +
  ggtitle("Symptom to Condition") +
  xlab("Symptom") +
  ylab("Count")
```

```r
library(Amelia)
missmap(train_long, main = "Missing values vs observed")
```

```r
model <- lm(value ~.,family=binomial(link='logit'),data=train_long)
```

```r
summary(model)
```

```r
anova(model, test="Chisq")
```

```r
# ROC Curve
library(pROC)
plot(model)
```

```r
library(caret)
library(glmnet)

# Elastic net regession
set.seed(123)
model <- train(
  value ~., data = train_long, method = "glmnet",
  trControl = trainControl("cv", number = 10),
  tuneLength = 10
)
# Best tuning parameter
model$bestTune
```

```r
# Model coefficients
coef(model$finalModel, model$bestTune$lambda)
```

```r
plot(model)
```
